```python
from GoogleNews import GoogleNews
import pandas as pd

def get_game_news():
    # Initialize the GoogleNews instance
    go_api = GoogleNews(lang='en', region='US', period='1d', encode='utf-8')

    # Search news articles related to games
    go_api.search('games')

    # Initialize an empty list to store the results
    results = []

    # Range of values
    for i in range(2, 20):
        go_api.getpage(i)
        results += go_api.result()

    # Create a DataFrame from the results
    outcome = pd.DataFrame(results)

    return outcome

# Call the function to get the game news
game_news = get_game_news()

# Display the first 20 rows
print(game_news.head(20))
```
