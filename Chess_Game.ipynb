```python
# Import all the libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Read the dataset
data = pd.read_csv('games.csv')

# Select features for clustering
x = data[['white_rating', 'black_rating']]

# Perform K-means clustering
kmeans = KMeans(n_clusters=4)  
# Fit the data x and predict the cluster labels for each data point
clusters = kmeans.fit_predict(x)

# Plot the clustered results
plt.scatter(x['white_rating'], x['black_rating'], c=clusters, cmap='rainbow')
plt.xlabel('White Rating')
plt.ylabel('Black Rating')
plt.title('K-means Clustering')
plt.show()
```
